# Model A (3 entires - proof of concept)

```python
from sklearn.linear_model import LinearRegression
```

```python
feature_columns =  master_sheet[["TDS", "Boron (B)", "Barium (Ba+2)", "Calcium (Ca+2)",	"Iron (Fe+2)", "Potassium (K+)", "Magnesium (Mg+2)", "Manganese (Mn+2)", "Sodium (Na+)", "Strontium (Sr+2)", "Zinc (Zn+2)", "Fluoride (F⁻)", "Chlorides (Cl-)",	"Nitrite (NO₂⁻)", "Bromide (Br⁻)", "Nitrate (NO₃⁻)", "Phosphate (PO₄³⁻)", "Sulfate (SO4-2)", "Bicarbonates (HCO3-)", "DR (R+)"]]
```

```python
target_columns = master_sheet[["500", "300", "250", "200", "150", "100", "50", "25", "12.5", "0"]]
```

```python
target_columns.head()
target_columns.shape
```

```python
X_train = feature_columns.loc[(feature_columns['TDS'] == 5876.0	) | (feature_columns['TDS']==164384.0)]
```

# Model B (removed 500 ppm variable)

```python
model_b.drop(columns=['500'])
model_b.dropna(inplace=True)
```

```python
feature_columns =  model_b[["TDS", "Boron (B)", "Barium (Ba+2)", "Calcium (Ca+2)",	"Iron (Fe+2)", "Potassium (K+)", "Magnesium (Mg+2)", "Manganese (Mn+2)", "Sodium (Na+)", "Strontium (Sr+2)", "Zinc (Zn+2)", "Fluoride (F⁻)", "Chlorides (Cl-)",	"Nitrite (NO₂⁻)", "Bromide (Br⁻)", "Nitrate (NO₃⁻)", "Phosphate (PO₄³⁻)", "Sulfate (SO4-2)", "Bicarbonates (HCO3-)"]]
```

```python
target_columns = model_b[["300", "250", "200", "150", "100", "50", "25", "12.5", "0"]]
```

```python
X_train = feature_columns.loc[(feature_columns['TDS'] == 5876.0	) | (feature_columns['TDS']==164384.0)]
```

```python
Y_train = target_columns.loc[X_train.index]
```

```python
X_test = feature_columns.loc[(feature_columns['TDS'] == 89081.0)]
```

```python
Y_test = target_columns.loc[X_test.index]
```

```python
reg = model.fit(X_train, Y_train)
reg.score(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
print(Y_pred)
```

### Insight:
A preliminary multi-output linear regression model was trained using standard curve data from two brines (5k and 164k TDS) and tested against an intermediate brine (~89k TDS). The model successfully reproduced the overall shape of the standard curve and generated absorbance predictions that were generally close to the observed values. The largest deviation occurred at the 500 ppm calibration point, which is also associated with reduced standard curve linearity (R² < 0.99) in laboratory measurements. Removing the 500 ppm point improved standard curve performance and suggests that the upper calibration range may contribute disproportionate variability.

```python
categorical_columns = master_sheet.select_dtypes(include=['object']).columns
encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(master_sheet[categorical_columns])
encoded_df = pd.DataFrame(encoded_data,
                          columns=encoder.get_feature_names_out(categorical_columns))
final_df = pd.concat([master_sheet.drop(columns=categorical_columns), encoded_df], axis = 1)
```

```python
final_df.head(15)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.drop(columns=['Friction Reducer_15,812', 'Friction Reducer_nan', 'Form_nan'], errors='ignore')
```

```python
fr_test = final_df['Friction Reducer_C'] == 1
fr_train= (final_df[(final_df['Friction Reducer_A'] == 1) | (final_df['Friction Reducer_B'] == 1)])
```

```python
feature_columns.head()
```
